## Generate code

### Subtask:
Generate code for a factorial function using three distinct prompts.


**Reasoning**:
Generate the factorial functions using three different prompts and store them in separate variables.



```python
# 1. Recursive implementation
recursive_prompt = "Generate a Python function to calculate the factorial of a number using recursion."
recursive_factorial_code = """
def recursive_factorial(n):
  if n == 0 or n == 1:
    return 1
  else:
    return n * recursive_factorial(n-1)
"""

# 2. Iterative implementation using a for loop
iterative_prompt = "Generate a Python function to calculate the factorial of a number using a for loop."
iterative_factorial_code = """
def iterative_factorial(n):
  result = 1
  for i in range(1, n + 1):
    result *= i
  return result
"""

# 3. Open-ended prompt
open_ended_prompt = "Generate a Python function to calculate the factorial of a number."
open_ended_factorial_code = """
def factorial(n):
  if n < 0:
    return "Factorial is not defined for negative numbers"
  elif n == 0:
    return 1
  else:
    result = 1
    for i in range(1, n + 1):
      result *= i
    return result
"""

print("Recursive Factorial Code:")
print(recursive_factorial_code)
print("\nIterative Factorial Code:")
print(iterative_factorial_code)
print("\nOpen-ended Factorial Code:")
print(open_ended_factorial_code)
```

```python
structured_comments = """
# Define a dictionary to simulate a user database.
# This dictionary will store user information with usernames as keys.
# Each value will be another dictionary containing user details,
# at a minimum, a 'password'.
users_db = {}

# Function: register_user
# Purpose: Registers a new user in the users_db.
# Inputs:
#   - username (str): The desired username for the new user.
#   - password (str): The password for the new user.
# Behavior:
#   - Check if the username already exists in users_db.
#   - If the username exists, return a message indicating that the user already exists.
#   - If the username does not exist, add the new user to users_db with the username as the key
#     and a dictionary containing the password as the value.
#   - Return a success message upon successful registration.

# Function: login_user
# Purpose: Authenticates an existing user.
# Inputs:
#   - username (str): The username of the user attempting to log in.
#   - password (str): The password provided by the user.
# Behavior:
#   - Check if the username exists in users_db.
#   - If the username does not exist, return a message indicating that the user was not found.
#   - If the username exists, compare the provided password with the stored password for that user.
#   - If the passwords match, return a success message.
#   - If the passwords do not match, return an invalid password message.
"""

linked_functions_code = """
# Define a dictionary to simulate a user database.
users_db = {}

def register_user(username, password):
  \"\"\"Registers a new user in the users_db.\"\"\"
  if username in users_db:
    return f"User '{username}' already exists."
  else:
    users_db[username] = {'password': password}
    return f"User '{username}' registered successfully."

def login_user(username, password):
  \"\"\"Authenticates an existing user.\"\"\"
  if username not in users_db:
    return f"User '{username}' not found."
  else:
    if users_db[username]['password'] == password:
      return f"User '{username}' logged in successfully."
    else:
      return "Invalid password."

# Example Usage:
print(register_user("alice", "secure_password"))
print(register_user("bob", "another_password"))
print(register_user("alice", "new_password")) # Attempt to register existing user

print(login_user("alice", "secure_password")) # Successful login
print(login_user("bob", "wrong_password"))    # Invalid password
print(login_user("charlie", "any_password"))  # User not found
"""

print("Structured Comments for Linked Functions:")
print(structured_comments)

print("\nGenerated Python Code for Linked Functions:")
print(linked_functions_code)
```
